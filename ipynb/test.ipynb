```python
import os, time
from dotenv import load_dotenv
from langchain_ollama import ChatOllama

load_dotenv() 


llm = ChatOllama(
    model=os.getenv("OLLAMA_MODEL"),
    base_url= os.getenv("OLLAMA_BASE_URL"),
    temperature=0.1,
    num_predict=200,   # límite de tokens de salida
)
```

```python
resp = llm.invoke("que tan bueno eres para crear codigo ?")
print(resp.content)
```

```python
for i in range(3):
    t0 = time.perf_counter()
    resp = llm.invoke("Di 'ok' y nada más.")
    dt = time.perf_counter() - t0
    print(i, f"{dt:.2f}s ->", resp.content)
```
